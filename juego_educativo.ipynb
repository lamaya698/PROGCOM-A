```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# Preguntas con opciones
preguntas = {
    "¬øCu√°nto es 5 + 3?": ["6", "7", "8", "9"],
    "¬øCu√°nto es 12 - 7?": ["3", "4", "5", "6"],
    "¬øCu√°nto es 6 * 4?": ["20", "24", "26", "30"],
    "¬øCu√°nto es 20 / 4?": ["3", "4", "5", "6"],
    "Si x + 2 = 10, ¬øcu√°nto vale x?": ["6", "7", "8", "9"],
    "¬øCu√°nto es el √°rea de un cuadrado de lado 5?": ["20", "25", "30", "35"],
    "Si un tri√°ngulo tiene base 10 y altura 4, ¬øcu√°l es su √°rea?": ["10", "15", "20", "25"]
}

respuestas_correctas = {
    "¬øCu√°nto es 5 + 3?": "8",
    "¬øCu√°nto es 12 - 7?": "5",
    "¬øCu√°nto es 6 * 4?": "24",
    "¬øCu√°nto es 20 / 4?": "5",
    "Si x + 2 = 10, ¬øcu√°nto vale x?": "8",
    "¬øCu√°nto es el √°rea de un cuadrado de lado 5?": "25",
    "Si un tri√°ngulo tiene base 10 y altura 4, ¬øcu√°l es su √°rea?": "20"
}

vidas_jugador = 3
vida_enemigo = 5
pregunta_actual = ""
opciones = []

def nueva_pregunta():
    """Genera una nueva pregunta aleatoria con opciones y la muestra."""
    global pregunta_actual, opciones

    if vidas_jugador == 0 or vida_enemigo == 0:
        clear_output(wait=True)
        if vida_enemigo == 0:
            display("üèÜ ¬°Ganaste! Has derrotado al enemigo. üéâ")
        else:
            display("üíÄ ¬°Perdiste! El enemigo te ha derrotado. ‚ò†Ô∏è")
        return

    pregunta_actual = random.choice(list(preguntas.keys()))
    opciones = preguntas[pregunta_actual]

    clear_output(wait=True)
    display("‚öîÔ∏è ¬°DUELO MATEM√ÅTICO! ‚öîÔ∏è")
    display(f"üî• Enemigo üíÄ Vida: {vida_enemigo}/5")
    display(f"üíñ Jugador ‚ù§Ô∏è Vidas: {vidas_jugador}/3")
    display("\nüß† Pregunta:")
    display(f"üëâ {pregunta_actual}")

    # Crear botones de opciones
    botones = [widgets.Button(description=op) for op in opciones]

    def on_button_click(b):
        verificar_respuesta(b.description)

    for boton in botones:
        boton.on_click(on_button_click)

    display(widgets.HBox(botones))

def verificar_respuesta(respuesta):
    """Verifica si la respuesta es correcta y actualiza el juego."""
    global vidas_jugador, vida_enemigo

    if respuesta == respuestas_correctas[pregunta_actual]:
        vida_enemigo -= 1
        display("‚úÖ ¬°Correcto! üí• El enemigo pierde 1 de vida.")
    else:
        vidas_jugador -= 1
        display(f"‚ùå Incorrecto. La respuesta era {respuestas_correctas[pregunta_actual]}. üò® Pierdes 1 vida.")

    nueva_pregunta()

# Iniciar el juego
nueva_pregunta()

```
