```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# Preguntas con opciones
preguntas = {
    "¿Cuánto es 5 + 3?": ["6", "7", "8", "9"],
    "¿Cuánto es 12 - 7?": ["3", "4", "5", "6"],
    "¿Cuánto es 6 * 4?": ["20", "24", "26", "30"],
    "¿Cuánto es 20 / 4?": ["3", "4", "5", "6"],
    "Si x + 2 = 10, ¿cuánto vale x?": ["6", "7", "8", "9"],
    "¿Cuánto es el área de un cuadrado de lado 5?": ["20", "25", "30", "35"],
    "Si un triángulo tiene base 10 y altura 4, ¿cuál es su área?": ["10", "15", "20", "25"]
}

respuestas_correctas = {
    "¿Cuánto es 5 + 3?": "8",
    "¿Cuánto es 12 - 7?": "5",
    "¿Cuánto es 6 * 4?": "24",
    "¿Cuánto es 20 / 4?": "5",
    "Si x + 2 = 10, ¿cuánto vale x?": "8",
    "¿Cuánto es el área de un cuadrado de lado 5?": "25",
    "Si un triángulo tiene base 10 y altura 4, ¿cuál es su área?": "20"
}

vidas_jugador = 3
vida_enemigo = 5
pregunta_actual = ""
opciones = []

def nueva_pregunta():
    """Genera una nueva pregunta aleatoria con opciones y la muestra."""
    global pregunta_actual, opciones

    if vidas_jugador == 0 or vida_enemigo == 0:
        clear_output(wait=True)
        if vida_enemigo == 0:
            display("🏆 ¡Ganaste! Has derrotado al enemigo. 🎉")
        else:
            display("💀 ¡Perdiste! El enemigo te ha derrotado. ☠️")
        return

    pregunta_actual = random.choice(list(preguntas.keys()))
    opciones = preguntas[pregunta_actual]

    clear_output(wait=True)
    display("⚔️ ¡DUELO MATEMÁTICO! ⚔️")
    display(f"🔥 Enemigo 💀 Vida: {vida_enemigo}/5")
    display(f"💖 Jugador ❤️ Vidas: {vidas_jugador}/3")
    display("\n🧠 Pregunta:")
    display(f"👉 {pregunta_actual}")

    # Crear botones de opciones
    botones = [widgets.Button(description=op) for op in opciones]

    def on_button_click(b):
        verificar_respuesta(b.description)

    for boton in botones:
        boton.on_click(on_button_click)

    display(widgets.HBox(botones))

def verificar_respuesta(respuesta):
    """Verifica si la respuesta es correcta y actualiza el juego."""
    global vidas_jugador, vida_enemigo

    if respuesta == respuestas_correctas[pregunta_actual]:
        vida_enemigo -= 1
        display("✅ ¡Correcto! 💥 El enemigo pierde 1 de vida.")
    else:
        vidas_jugador -= 1
        display(f"❌ Incorrecto. La respuesta era {respuestas_correctas[pregunta_actual]}. 😨 Pierdes 1 vida.")

    nueva_pregunta()

# Iniciar el juego
nueva_pregunta()

```
