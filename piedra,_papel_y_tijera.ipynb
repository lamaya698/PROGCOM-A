```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

def play(user_choice):
    choices = ['Piedra', 'Papel', 'Tijera']
    computer_choice = random.choice(choices)

    if user_choice == computer_choice:
        result = "¡Es un empate!"
    elif (user_choice == "Piedra" and computer_choice == "Tijera") or \
         (user_choice == "Papel" and computer_choice == "Piedra") or \
         (user_choice == "Tijera" and computer_choice == "Papel"):
        result = "¡Ganaste!"
    else:
        result = "¡Perdiste!"

    clear_output()
    display_game()

    images_ascii = {
        "Piedra": """
        ******
       *    *
      *      *
      *      *
       *    *
        ******
        """,
        "Papel": """
        **********
        *        *
        *        *
        *        *
        **********
        """,
        "Tijera": """
        *    *
         *  *
          **
         *  *
        *    *
        """
    }

    print(f"Tú elegiste: {user_choice}")
    print(images_ascii[user_choice])
    print(f"La computadora eligió: {computer_choice}")
    print(images_ascii[computer_choice])
    print(result)

def display_game():
    print("Elige una opción:")

    button_piedra = widgets.Button(description="Piedra")
    button_papel = widgets.Button(description="Papel")
    button_tijera = widgets.Button(description="Tijera")

    button_piedra.on_click(lambda _: play("Piedra"))
    button_papel.on_click(lambda _: play("Papel"))
    button_tijera.on_click(lambda _: play("Tijera"))

    display(button_piedra, button_papel, button_tijera)

display_game()
```
